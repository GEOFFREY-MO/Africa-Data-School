satellite dataset

```python
import pandas as pd
#Numpy here is a numerical python useful for numerical culculations
import numpy as np
#importing pyplot module frommatoplotlib library
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans


#Let now load the data from the text file using pandas library
df = pd.read_csv('satellite_training.txt',sep='\s+',header=None)
# Since our last column in our dataset contain different features we need # to extract the feature data from all columns except the last column
X = df.iloc[:, :-1].values


# Initializing of K-means using random partition method as recommended in    # the task
def initialize_clusters(X, k):
    clusters=np.zeros(len(X))
    for i in range(len(X)):
        clusters[i]=np.random.randint(0,k)
    return clusters

# Now here computing the initial centroids
def compute_centroids(X, clusters, k):
    centroids= np.zeros((k, X.shape[1]))
    for i in range(k):
        if np.isnan(np.sum(X[clusters == i])):
            # If the slice is empty the previous centroid is kept
            centroids[i]=centroids[i-1]
        else:
            centroids[i]=np.mean(X[clusters==i], axis=0)
    return centroids


#Then computation of  the sum of all squared errors (SSE)in this method
def compute_sse(X, clusters, centroids):
    sse = 0
    for i in range(len(X)):
        sse += np.linalg.norm(X[i]- centroids[int(clusters[i])])**2
    return sse

#execute K-means in the a range of K values (2-10)as spacefied using a for #loop
k_values = range(2, 11)
sse_values = []

for k in k_values:
    clusters = initialize_clusters(X, k)
    for _ in range(20):
        centroids = compute_centroids(X, clusters, k)
        clusters = np.apply_along_axis(lambda x: np.argmin([np.linalg.norm(x - c) for c in centroids]), 1, X)
    sse = compute_sse(X, clusters, centroids)
    sse_values.append(sse)

# let now display the SSE values in four decimal places
for k, sse in zip(k_values, sse_values):
    print("K = {}: SSE = {:.4f}".format(k, sse))



# lastly in pyplot module from matplotlib library plot the k vs SSE values
plt.plot(k_values, sse_values)
plt.xlabel('K')
plt.ylabel('SSE')
plt.title('SSE vs.K for satellite')
plt.show()
```
